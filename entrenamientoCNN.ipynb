```python
# Importar bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
import random
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

from tqdm import tqdm  # Para barra de progreso (opcional)

from sklearn.model_selection import train_test_split
from tensorflow.keras.regularizers import l2
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from functools import partial
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import ResNet50
from sklearn.model_selection import StratifiedKFold
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.applications.resnet50 import preprocess_input

'''
# Verificar si estamos usando GPU
from tensorflow.python.client import device_lib
print(device_lib.list_local_devices())
'''

'''
# Descargar el dataset (en Colab)
!wget https://www.dropbox.com/s/opuvvdv3uligypx/fer2013.tar.gz
!tar -xzf fer2013.tar.gz
!rm fer2013.tar.gz
'''
#---------------------------------------------------------------------------
# Función de balanceo de clases
'''
¿Por qué?:

* El dataset original tiene desbalanceo (muchos "Happy", pocos "Disgust")
* Limitar muestras por clase evita dominancia de clases mayoritarias
* Mejora la capacidad de generalización para todas las emociones

'''
#----------------------------------------------------------------------------


def balance_classes(X, y, max_samples=300, min_samples=100):
    """Balanceo agresivo con mínimos y máximos estrictos"""
    unique_classes, counts = np.unique(y, return_counts=True)
    balanced_X, balanced_y = [], []

    for cls in unique_classes:
        indices = np.where(y == cls)[0]
        # Forzar mínimo de muestras (oversampling)
        if len(indices) < min_samples:
            indices = np.random.choice(indices, min_samples, replace=True)  # Oversampling
        # Limitar máximo de muestras
        elif len(indices) > max_samples:
            indices = np.random.choice(indices, max_samples, replace=False)

        balanced_X.extend(X[indices])
        balanced_y.extend(y[indices])

    return np.array(balanced_X), np.array(balanced_y)

#----------------------------------------------------------
#Función para cargar datos con porcentajes personalizables
#----------------------------------------------------------

def load_fer2013_with_percentage(base_path='fer2013',
                                train_percentage=0.1,
                                test_percentage=0.1,
                                train_split=0.6,
                                target_size=(224, 224),
                                balance=True,
                                max_samples=500):

    X, y = [], []

    # Diccionario de emociones
    emotion_dict = {'angry': 0, 'disgust': 1, 'fear': 2,
                    'happy': 3, 'neutral': 4, 'sad': 5, 'surprise': 6}

    # Cargar imágenes de entrenamiento
    train_path = os.path.join(base_path, 'train') #ruta train
    test_path = os.path.join(base_path, 'test') #ruta

    # Procesar train y test por separado
    X_train, y_train = _load_from_folder(train_path, emotion_dict, train_percentage, target_size)
    X_test, y_test = _load_from_folder(test_path, emotion_dict, test_percentage, target_size)

    # Combinar y dividir según el train_split
    X_combined = np.concatenate((X_train, X_test))
    y_combined = np.concatenate((y_train, y_test))

    # Dividir según el porcentaje deseado
    X_train_final, X_test_final, y_train_final, y_test_final = train_test_split(
        X_combined, y_combined,
        train_size=train_split,
        stratify=y_combined,
        random_state=42
    )
# Mejoras de balanceo
    # Balancear clases si se solicita
    if balance:
            X_train_final, y_train_final = balance_classes(X_train_final, y_train_final, max_samples)
            X_test_final, y_test_final = balance_classes(X_test_final, y_test_final, max_samples)

    # Verificar distribución final
    print("\nDistribución final:")
    unique_train, counts_train = np.unique(y_train, return_counts=True)
    unique_test, counts_test = np.unique(y_test, return_counts=True)

    for cls, count in zip(unique_train, counts_train):
        print(f"Train - Clase {cls}: {count} muestras")

    for cls, count in zip(unique_test, counts_test):
        print(f"Test - Clase {cls}: {count} muestras")


    return X_train_final, X_test_final, y_train_final, y_test_final


#-----------------------------------------
# Mejoras en la Carga y Limpieza de Datos
#-----------------------------------------

# En _load_from_folder agrego min_samples=20 para garantizar mínimo 20 muestras por etiqueta

def _load_from_folder(folder_path, emotion_dict, percentage, target_size, min_samples=20):
    X, y = [], []
    for emotion, label in emotion_dict.items():
        emotion_folder = os.path.join(folder_path, emotion)
        if not os.path.exists(emotion_folder):
            continue

        all_files = [f for f in os.listdir(emotion_folder)
                    if f.endswith(('.jpg', '.jpeg', '.png'))]  # Solo archivos de imagen

        # Verificar mínimo de muestras
        if len(all_files) < min_samples:
            print(f"Advertencia: Muy pocas muestras ({len(all_files)}) para clase {emotion}")
            continue

        num_files = max(min_samples, int(len(all_files) * percentage))  # Al menos min_samples
        selected_files = random.sample(all_files, num_files)

        for file_name in tqdm(selected_files, desc=f'Cargando {emotion}'):
            try:
                img_path = os.path.join(emotion_folder, file_name)
                img = cv2.imread(img_path)
                if img is None:  # Verificar si la imagen se cargó correctamente
                    continue

                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

                # Mejorar contraste (CLAHE)
                lab = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)
                l, a, b = cv2.split(lab)
                clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
                cl = clahe.apply(l)
                limg = cv2.merge((cl,a,b))
                img = cv2.cvtColor(limg, cv2.COLOR_LAB2RGB)

                img = cv2.resize(img, target_size)

                # Verificar imágenes corruptas o vacías
                if np.mean(img) < 10:  # Imagen casi vacía
                    continue

                X.append(img)
                y.append(label)

            except Exception as e:
                print(f"Error procesando {file_name}: {str(e)}")
                continue

    return np.array(X), np.array(y)


#-----------------------------------------
#Cargar datos con porcentajes específicos
#-----------------------------------------

# Configuración de porcentajes
total_data_percentage = 0.1  # Usar solo 10% del dataset completo
train_split = 0.6           # Del 10%, 60% para train y 40% para test

# Cargar datos
X_train, X_test, y_train, y_test = load_fer2013_with_percentage(
    train_percentage=total_data_percentage,
    test_percentage=total_data_percentage,
    train_split=train_split
)

# Mostrar estadísticas
print(f"Total de imágenes cargadas: {len(X_train) + len(X_test)}")
print(f"Train: {len(X_train)} imágenes ({len(X_train)/(len(X_train)+len(X_test))*100:.1f}%)")
print(f"Test: {len(X_test)} imágenes ({len(X_test)/(len(X_train)+len(X_test))*100:.1f}%)")
print("Distribución de clases en train:", np.bincount(y_train))
print("Distribución de clases en test:", np.bincount(y_test))

# Visualizar ejemplos
plt.figure(figsize=(12, 6))
for i in range(10):
    plt.subplot(2, 5, i+1)
    plt.imshow(X_train[i])
    plt.title(f"Label: {y_train[i]}")
    plt.axis('off')
plt.tight_layout()
plt.show()


# -----------------------------------------
# Visualización de muestras por clase
# -----------------------------------------

def visualize_samples(X, y, class_names, samples=5):
    plt.figure(figsize=(15, 10))
    y_int = np.argmax(y, axis=1) if len(y.shape) > 1 else y

    for class_id, class_name in enumerate(class_names):
        indices = np.where(y_int == class_id)[0][:samples]

        for i, idx in enumerate(indices):
            plt.subplot(len(class_names), samples, class_id * samples + i + 1)
            plt.imshow(X[idx].astype('uint8'))
            plt.title(f"{class_name}-{idx}")
            plt.axis('off')

    plt.tight_layout()
    plt.show()

# Convertir a one-hot primero
y_train = to_categorical(y_train, num_classes=7)
y_test = to_categorical(y_test, num_classes=7)

# Llamar a la función (antes del entrenamiento)
class_names = ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
visualize_samples(X_train, y_train, class_names)

# Opcional: Eliminar muestras problemáticas manualmente
# bad_samples = [123, 456]  # Ejemplo: IDs de imágenes mal etiquetadas
# X_train = np.delete(X_train, bad_samples, axis=0)
# y_train = np.delete(y_train, bad_samples, axis=0)



#-----------------------------------------
#Preprosesamiento de imagen para resNet50
#-----------------------------------------

'''
# Aplicar antes de preprocess_input
X_train = enhanced_preprocessing(X_train)
X_test = enhanced_preprocessing(X_test)

# Preprocesamiento específico para ResNet50
X_train = preprocess_input(X_train)
X_test = preprocess_input(X_test)

# Convertir etiquetas a one-hot encoding
y_train = to_categorical(y_train, num_classes=7)
y_test = to_categorical(y_test, num_classes=7)
'''

# -----------------------------------------
# Preprocesamiento especializado para rostros
# -----------------------------------------
def enhanced_face_preprocessing(images, y=None):
    """Pipeline mejorado con detección facial"""
    processed = []
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

    for img in images:
        try:
            # 1. Detección de rostro
            gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
            faces = face_cascade.detectMultiScale(gray, 1.1, 4)

            if len(faces) > 0:
                x, y, w, h = faces[0]
                img = img[y:y+h, x:x+w]  # Recortar ROI

            # 2. Mejora de contraste LAB
            img_lab = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)
            l, a, b = cv2.split(img_lab)
            clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
            img_lab = cv2.merge((clahe.apply(l), a, b))
            img = cv2.cvtColor(img_lab, cv2.COLOR_LAB2RGB)

            # 3. Redimensionamiento final
            img = cv2.resize(img, (224, 224))
            processed.append(img)

        except Exception as e:
            print(f"Error en preprocesamiento: {str(e)}")
            processed.append(img)  # Conservar original si falla

    return np.array(processed)

# Aplicar preprocesamiento
print("Aplicando preprocesamiento facial...")
X_train = enhanced_face_preprocessing(X_train)
X_test = enhanced_face_preprocessing(X_test)

# Preprocesamiento específico de ResNet50
X_train = preprocess_input(X_train)
X_test = preprocess_input(X_test)


## Data augmentation mejorado ##
# -----------------------------------------
# Data Augmentation por clase
# -----------------------------------------

# Configurar generador
# 1. Primero definimos una función de preprocesamiento combinada
def combined_augmentation(image):
    """Aplica aumentos generales + específicos por clase"""
    # Aumentos básicos (ya incluidos en ImageDataGenerator)
    image = image + np.random.normal(0, 0.03, image.shape)  # Ruido gaussiano

    # Convertir a uint8 para operaciones CV
    image_uint8 = np.clip(image, 0, 255).astype('uint8')

    # Aumentos especiales para disgust (clase 1)
    if np.random.rand() > 0.7:  # Aplicar solo el 30% de las veces
        if np.argmax(y_train[np.random.randint(0, len(y_train))]) == 1:  # Check clase disgust
            image_uint8 = cv2.rotate(image_uint8, cv2.ROTATE_90_CLOCKWISE)
            image_uint8 = cv2.flip(image_uint8, 1)

    return image_uint8.astype('float32')

# 2. Configurar generador
datagen = ImageDataGenerator(
    rotation_range=25,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.15,
    zoom_range=0.3,
    horizontal_flip=True,
    brightness_range=[0.6, 1.4],
    fill_mode='reflect',
    preprocessing_function=combined_augmentation  # Usamos nuestra función combinada
)

# 3. Crear generador (CORRECTO)
train_generator = datagen.flow(
    X_train,
    y_train,
    batch_size=32,
    shuffle=True,
    seed=42  # Añadir semilla para reproducibilidad
)

#-----------------------------------------
#Crear modelo basado en resNet50
#-----------------------------------------


def create_enhanced_resnet50(input_shape=(224, 224, 3), num_classes=7):
  # Cargar ResNet50 sin la capa superior (include_top=False)
    base_model = ResNet50(weights='imagenet',
                          include_top=False,
                          input_shape=input_shape)

    '''
    # sin fine-tunning
    # Congelar las capas base (opcional, para fine-tuning)
    for layer in base_model.layers:
        layer.trainable = False
    '''

    # Descongelar capas superiores (layer.trainable = True / con fine-tuning)
    for layer in base_model.layers[-20:]:
        layer.trainable = True

    # Añadir nuestras propias capas

    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    #512 neuronas
    x = Dense(512, activation='relu', kernel_regularizer=l2(0.01))(x)
    x = Dropout(0.6)(x)
    #256 neuronas
    x = Dense(256, activation='relu', kernel_regularizer=l2(0.01))(x)
    x = Dropout(0.5)(x)

    predictions = Dense(num_classes, activation='softmax')(x)

    # Modelo final
    model = Model(inputs=base_model.input, outputs=predictions)

    return model


# Crear modelo
model = create_enhanced_resnet50()
model.summary()


#-----------------------------------------
#Compilar y entrenar el modelo
#-----------------------------------------


# -----------------------------------------
# Entrenamiento con Validación Cruzada
# -----------------------------------------

skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

# Verificar si y_train ya es one-hot
if len(y_train.shape) == 1:  # Si no es one-hot
    y_train_int = y_train
else:  # Si ya es one-hot
    y_train_int = np.argmax(y_train, axis=1)

for fold, (train_idx, val_idx) in enumerate(skf.split(X_train, y_train_int)):
    print(f"\n{'='*50}")
    print(f"Entrenando fold {fold + 1}/3")
    print(f"{'='*50}")

    # Crear modelo nuevo para cada fold
    model = create_enhanced_resnet50()
    model.compile(optimizer=Adam(learning_rate=0.0001),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])



 # Callbacks por fold
callbacks = [
        EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True),
        ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5),
        ModelCheckpoint(f'best_model_fold{fold}.keras', save_best_only=True)
      ]

    # Datos para este fold
X_fold_train, X_fold_val = X_train[train_idx], X_train[val_idx]
y_fold_train, y_fold_val = y_train[train_idx], y_train[val_idx]

    # Entrenamiento

history = model.fit(
    datagen.flow(
        x=X_fold_train,
        y=y_fold_train,
        batch_size=32,
        shuffle=True,
        seed=42
    ),
    epochs=50,
    validation_data=(X_fold_val, y_fold_val),
    callbacks=callbacks,
    verbose=1  # Añadir esto para ver progreso
)

    # Evaluar en test
test_loss, test_acc = model.evaluate(X_test, y_test, verbose=0)
print(f"\nPrecisión en test (fold {fold}): {test_acc:.4f}")

# -----------------------------------------
# Análisis de Errores Post-Entrenamiento
# -----------------------------------------



def analyze_results(model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_pred_classes = np.argmax(y_pred, axis=1)
    y_true = np.argmax(y_test, axis=1)

    # Reporte de clasificación
    print("\nClassification Report:")
    print(classification_report(y_true, y_pred_classes, target_names=class_names))

    # Matriz de confusión
    plt.figure(figsize=(10,8))
    cm = confusion_matrix(y_true, y_pred_classes)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Matriz de Confusión')
    plt.ylabel('Verdadero')
    plt.xlabel('Predicho')
    plt.show()

# Analizar resultados del último fold
    analyze_results(model, X_test, y_test)
```
